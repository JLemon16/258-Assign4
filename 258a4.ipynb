```python
# Data generation 
 
import numpy as np
import math as mt
import matplotlib.pyplot as plt
import random
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

 
x=np.linspace(-2*mt.pi,2*mt.pi,1000)
 
ya=(np.sin(x))*x**2+2*np.sin(x)
 
z=np.random.randn(1,1000)
 
y =ya+z
 
X_data=np.reshape(x,[1000,1])
 
y_data=np.reshape(y,[1000,1])
 
from keras.models import Sequential
 
from keras.layers import Dense
 
mymodel = Sequential()
 
# input and Hidden layer-1
 
mymodel.add(Dense(units=50, input_dim=1,
kernel_initializer='normal',activation='relu'))
 
# Hidden layer-2
 
mymodel.add(Dense(units=50,
kernel_initializer='normal',activation='relu'))

# Hidden layer-3
 
mymodel.add(Dense(units=50,
kernel_initializer='normal',activation='relu'))

 
# output layer
 
mymodel.add(Dense(units=1, kernel_initializer='normal'))
 
print(mymodel.summary())
 
from sklearn.model_selection import KFold
 
sum_mse=0
 
sum_r_square=0
 
kf=KFold(n_splits=5)
 
for train_index, test_index in kf.split(X_data):
 
    X_train, X_test = X_data[train_index], X_data[test_index]
 
    y_train, y_test = y_data[train_index], y_data[test_index]
 
    # Training our model build above

    mymodel.compile(loss='mean_squared_error', optimizer='adam')
    # Training model

    mymodel.fit(X_train, y_train, batch_size = 32, epochs = 50,
verbose=0)
    # predicting output of trained model using test input data X_test
 
    y_pred=mymodel.predict(X_test)
 
    # MSE calculation base true output and model output
 
    mse = mean_squared_error(y_test, y_pred)

    # R-square calculation base true output and model output
 
    r_squared = r2_score(y_test, y_pred)
 
    # MSE calculation base true output and model output
 
    mse = mean_squared_error(y_test, y_pred)
 
    # R-square calculation base true output and model output
 
    r_squared = r2_score(y_test, y_pred)
 
    sum_mse=+mse
 
    sum_r_square=+r_squared
avg_mse=sum_mse/5
 
avg_r_squared= sum_r_square /5
 
print(f'mse = {avg_mse:.2}')
 
print(f'r_squared = {avg_r_squared:.2%}')
```
